```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('nba.csv')
# Descriptive Statistics
print(df.describe())

# Print the first five rows of the data
print(df.head())

# Print the number of rows and columns
print(f"Rows: {df.shape[0]}")

# Print the column names
print(df.columns)

# visualize the players' PER(per salary one year)
df['PER'].hist()
plt.title('PER Distribution')
plt.xlabel('PER')
plt.ylabel('Number of Players')
plt.show()
```

```python
# Converting continuous variables (quantitative variables) in the data to discrete variables
df['PTS'] = pd.cut(df['PTS'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['FGM'] = pd.cut(df['FGM'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['AST'] = pd.cut(df['AST'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['REB'] = pd.cut(df['REB'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['FTM'] = pd.cut(df['FTM'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['FGA'] = pd.cut(df['FGA'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['FTA'] = pd.cut(df['FTA'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['STL'] = pd.cut(df['STL'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
df['BLK'] = pd.cut(df['BLK'], bins=5, labels=['low', 'medium', 'high',"very high","extreme"])
# visualize these variables in a box plot
df.boxplot(column=['PER'], by='PTS', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='FGM', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='AST', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='REB', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='FTM', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='FGA', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='FTA', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='STL', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
df.boxplot(column=['PER'], by='BLK', figsize=(10, 10))
plt.ylabel('PER')
plt.show()
```
